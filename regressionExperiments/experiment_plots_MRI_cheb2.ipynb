```python
import torch
```

```python
# direct reconstruction experiments
deg_Quads = [40, 50, 60, 65, 70, 80]

```

```python
LSTQS_direct = []
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure
```

```python
Analytical_direct = torch.load('/media/chethan/New Volume/Thesis/regressionExperiments/cheb2_Analytic_MRI_direct_recon_psnr.pt')
RK_direct = torch.load('/media/chethan/New Volume/Thesis/regressionExperiments/cheb2_RK_MRI_direct_recon_psnr.pt')
LSTQS_direct = torch.load('/media/chethan/New Volume/Thesis/regressionExperiments/cheb2_LSTQS_MRI_direct_recon_psnr.pt')
Sobolev_direct = torch.load('/media/chethan/New Volume/Thesis/regressionExperiments/cheb2_sobolev_MRI_direct_recon_psnr.pt')

figure(figsize=(8, 6), dpi=80)
plt.plot(deg_Quads, Analytical_direct, label = 'Unperturbed Polynomial Evaluation Analytical')
plt.plot(deg_Quads, RK_direct, label = 'Unperturbed Polynomial Evaluation Runge Kutta')
plt.plot(deg_Quads, LSTQS_direct, label = 'Unperturbed Polynomial Evaluation LSTQS')
plt.plot(deg_Quads, Sobolev_direct, label = 'Unperturbed Polynomial Evaluation Sobolev')

plt.legend()

plt.xticks(deg_Quads, fontsize=12 ) 
plt.yticks(fontsize= 12) 
plt.xlabel('Legendre Polynomial degree',fontsize=15)
plt.ylabel('PSNR',fontsize=15)
```
